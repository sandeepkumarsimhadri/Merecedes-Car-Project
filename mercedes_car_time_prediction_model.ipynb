```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
mc_df=pd.read_csv('Mercedes (1).csv')
mc_df.shape
```

```python
mc_df['y']
```

```python
cat_cols=[col for col in mc_df.columns if mc_df[col].dtype=='object']## Finding catg cols
cat_cols
```

```python
mc_df['X1'].value_counts()
```

```python
[mc_df[col].unique() for col in cat_cols]# Finding unique catg in each catg column
```

```python
y=mc_df['y']
#mc_df.drop(columns=['y','ID'],inplace=True)
con_cols=[col for col in mc_df.columns if mc_df[col].dtype=='int64' or mc_df[col].dtype=='float64']
# Finding cont cols
```

```python
[mc_df[col].value_counts() for col in con_cols]# finding values in cont cols and found only two catg
```

```python
tot_cols=con_cols+cat_cols
mc_df.columns,len(tot_cols)
```

```python
t_df=mc_df.copy()
t_df.drop(columns=cat_cols,inplace=True)
```

```python
temp_df=pd.DataFrame({'cols':t_df.columns,'var':np.var(mc_df[con_cols])})
one_var_catg_cols=list(temp_df[temp_df['var']==0.0]['cols'])
```

```python
for col in con_cols:
    mc_df[col]=mc_df[col].astype(object)
# converting all cont cols to catg cols
```

```python
cat_cols=[col for col in mc_df.columns if mc_df[col].dtype=='object']
len(cat_cols)
```

```python
[col for col in mc_df.columns if mc_df[col].dtype=='int64' or mc_df[col].dtype=='float64']
```

```python
mc_df['X93'].value_counts()
# single value catg col
```

```python
cols_drop_uniq_val=[]
for col in con_cols:
    if len(mc_df[col].value_counts()) ==1 or len(mc_df[col].value_counts())==mc_df.shape[0]:
        cols_drop_uniq_val.append(col)
cols_drop_uniq_val
```

```python
mc_df.drop(columns=one_var_catg_cols,inplace=True)
# droping all single value catg cols
```

```python
mc_df.drop(columns=['y','ID'],inplace=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(mc_df,y,random_state=63,test_size=0.2)
```

```python
[col for col in mc_df.columns if mc_df[col].dtype=='int64' or mc_df[col].dtype=='float64']
```

```python
ful_cat_cols=[col for col in mc_df.columns if mc_df[col].dtype=='object']
len(ful_cat_cols)
```

```python
x_train[ful_cat_cols].isnull().sum().any()## no missing values
```

```python
cat_encd_train= pd.get_dummies(x_train[ful_cat_cols])
cat_encd_test=pd.get_dummies(x_test[ful_cat_cols])
```

```python
cat_encd_train.shape,cat_encd_test.shape
```

```python
cat_encd_train_final,cat_encd_test_final=cat_encd_train.align(cat_encd_test,join='inner',axis=1)
cat_encd_train_final.shape###aligning train & test data one hot encoded catg columns due to unqual no of columns i.e no of cilumns would differ for that we align to get same
```

```python
dtree_reg=DecisionTreeRegressor()
```

```python
dtree_reg.fit(cat_encd_train_final,y_train)
```

```python
y_test_pred=dtree_reg.predict(cat_encd_test_final)
y_test_pred
```

```python
dtree_reg.score(cat_encd_test_final,y_test)
```

```python
dtree_reg.score(cat_encd_train_final,y_train)
```
